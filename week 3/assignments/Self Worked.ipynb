```python
# Answer Eleven
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
cont=new_df
data=pd.DataFrame({"Country":ContinentDict.keys(),"Continent":ContinentDict.values()}).set_index("Country")
new=pd.merge(cont,data,how="left", on="Country")
# new.drop("Continent_x", axis=1, inplace=True)
#new.rename(columns={"Continent_y":"Continent"}, inplace=True)
new[["size","sum","mean","std"]]=None
new=new.set_index('Continent')
new=new.groupby("Continent")["Pop"].agg(size=np.size,sum=np.sum,mean=np.mean,std=np.std)
new
```
