# Splitting

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/census.csv')
# And exclude state level summarizations, which have sum level value of 40
df = df[df['SUMLEV']==50]
df.head()
```

```python
# Pandas dataframe.groupby() function is used to split the data into groups based on some criteria. 
# Pandas objects can be split on any of their axes. 
# The abstract definition of grouping is to provide a mapping of labels to group names.
# Groupby() returns a tuple, where the first value is the value of the key we were trying to group by, 
# in this case a specific state name, and the second one is projected dataframe that was found for that group.
# It group label values as unique values and their store their subsequent values in a frame.

states = df.groupby('STNAME')
states.first()
```

```python
states.get_group('Alabama')
```

```python
for group, frame in df.groupby('STNAME'):
    print(group,frame['CTYNAME'])
   #  avg = np.average(frame['CENSUS2010POP'])
    # And print the results
    # print('Counties in state ' + group + 
     #     ' have an average population of ' + str(avg))
```

# Applying

```python
# To this point we have applied very simple processing to our data after splitting, really just outputting
# some print statements to demonstrate how the splitting works. The pandas developers have three broad
# categories of data processing to happen during the apply step, Aggregation of group data, Transformation of
# group data, and Filtration of group data
```

## Aggregation

```python
# The agg() method allows you to apply a function or a list of function names to be executed along one of the axis 
# of the DataFrame, default 0, which is the index (row) axis.
# The most straight forward apply step is the aggregation of data, and uses the method agg() on the groupby object.
# With agg we can pass in a dictionary of the columns we are interested in aggregating along with the function we 
# are looking to apply to aggregate.
df=pd.read_csv("datasets/listings.csv")
df.head()
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value":np.average})
# it doesn't work because np.average does not ignore nans! However, there is a function we can use for this
```

```python
df.groupby("cancellation_policy").agg({"review_scores_value":np.nanmean})
# It ignores NaN values.
```

```python
# We can just extend this dictionary to aggregate by multiple functions or multiple columns.
df.groupby("cancellation_policy").agg({"review_scores_value":(np.nanmean,np.nanstd),
                                      "reviews_per_month":np.nanmean})
```

## Transformation

```python
# The transform() method allows you to execute a function for each value of the DataFrame.
# it broadcasts the function you supply over the grouped dataframe, returning a new dataframe. 
# This makes combining data later easy.
```

```python
cols=['cancellation_policy','review_scores_value']
# Now lets transform it, I'll store this in its own dataframe
transform_df=df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
# It returns a series whose length is is equal to our actual dataframe, so that we can also merge it in old dataframe.
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'}, axis='columns', inplace=True)
df=df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

## Filtering

```python
# The filter() method filters the DataFrame, and returns only the rows or columns that are specified in the filter.
# The filter() function takes in a function which it applies to each group dataframe and returns either a True or a False,
# depending upon whether that group should be included in the results.
```

```python
# For instance, if we only want those groups which have a mean rating above 9 included in our results
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9)
```
